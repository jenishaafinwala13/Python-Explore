# Conditional Statements

Syntax of If condition

if condition:
    if-statments

```python
num = 10
if num % 2 == 0:
    print("num is even.")
```

 if…else statement

 if condition:
    if-block
else:
    else-block

```python
num = 10
if num % 2 != 0:
    print("num is odd.")
else:
    print("num is even.")
```

if…elif…else statement

if if-condition:
    if-block
elif elif-condition1:
    elif-block1
elif elif-condition2:
    elif-block2
...
else:
    else-block

```python
num = 11
if num % 2 != 0:
    print("num is odd.")
elif num == 0:
    print("num is zero.")
else:
    print("num is even.")
```

# Looping Statement

For Loop

```python
words = ['cat', 'window', 'home', 'password']

for w in words:
    print(w, len(w))
```

While Loop
while

```python
a, n = 0, 10
while a < n:
    print(a)
    a = a + 1
```
