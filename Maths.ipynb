Triangular Numbers
The formula for the sum of integers from 1 to N
 (also known as triangular numbers) is given by:


```python
N = 1000
left_side = sum([k for k in range(1,N+1)])
right_side = N*(N+1)/2
print(left_side)
print(right_side)
```
