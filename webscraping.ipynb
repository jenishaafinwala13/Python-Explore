```python
import requests
from bs4 import BeautifulSoup

def scrape_indeed_jobs(query, location):
    base_url = "https://in.indeed.com/jobs?"
    params = {
        'q': query,
        'l': location,
    }

    # Send a GET request to Indeed with the specified parameters
    response = requests.get(base_url, params=params)

    if response.status_code == 200:
        # Parse the HTML content of the page
        soup = BeautifulSoup(response.content, 'html.parser')

        # Find job postings on the page
        job_posts = soup.find_all('div', class_='jobsearch-SerpJobCard')

        for job_post in job_posts:
            # Extract relevant information from each job posting
            job_title = job_post.find('h2', class_='title').a.get('title')
            company = job_post.find('span', class_='company').text.strip()
            location = job_post.find('div', class_='recJobLoc').get('data-rc-loc')
            link = base_url + job_post.find('h2', class_='title').a.get('href')

            print(f"Job Title: {job_title}")
            print(f"Company: {company}")
            print(f"Location: {location}")
            print(f"Link: {link}")
            print("\n---\n")

    else:
        print(f"Failed to retrieve page. Status code: {response.status_code}")

# Example usage for searching Python jobs in India
scrape_indeed_jobs(query='python', location='surat')
```

```python

import re
import json
import requests
from urllib.parse import urlencode

def get_indeed_search_url(keyword, location, offset=0):
    parameters = {"q": keyword, "l": location, "filter": 0, "start": offset}
    return "https://in.indeed.com/jobs?" + urlencode(parameters)


headers={"User-Agent": "Mozilla/5.0 (iPad; CPU OS 12_2 like Mac OS X) AppleWebKit/605.1.15 (KHTML, like Gecko) Mobile/15E148"}

job_id_list = []

## Job Search Parameters
keyword_list = ['software engineer']
location_list = ['India']

## Loop Through Indeed Pages Until No More Jobs
for keyword in keyword_list:
    for location in location_list:
        for offset in range(0, 1010, 10):
            try:
                indeed_jobs_url = get_indeed_search_url(keyword, location, offset)
                response = requests.get(indeed_jobs_url, headers=headers)

                if response.status_code == 200:
                    script_tag  = re.findall(r'window.mosaic.providerData\["mosaic-provider-jobcards"\]=(\{.+?\});', response.text)
                    if script_tag is not None:
                        json_blob = json.loads(script_tag[0])
                        jobs_list = json_blob['metaData']['mosaicProviderJobCardsModel']['results']
                        for index, job in enumerate(jobs_list):
                            if job.get('jobkey') is not None:
                                job_id_list.append(job.get('jobkey'))

                        ## If response contains less than 10 jobs then stop pagination
                        if len(jobs_list) < 10:
                            break

            except Exception as e:
                print('Error', e)

print(job_id_list)
```

```python
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.chrome.options import Options

def scrape_indeed_jobs(query, location):
    # Set up the WebDriver with headless option
    options = Options()
    options.add_argument('--headless')
    driver = webdriver.Chrome(options=options)

    # Define the Indeed URL with the query and location
    url = f'https://in.indeed.com/jobs?q={query}&l={location}'

    # Open the Indeed website
    driver.get(url)

    # Find job postings on the page
    job_posts = driver.find_elements_by_class_name('jobsearch-SerpJobCard')

    for job_post in job_posts:
        # Extract relevant information from each job posting
        job_title = job_post.find_element_by_class_name('title').find_element_by_tag_name('a').get_attribute('title')
        company = job_post.find_element_by_class_name('company').text.strip()
        location = job_post.find_element_by_class_name('location').text.strip()
        link = job_post.find_element_by_class_name('title').find_element_by_tag_name('a').get_attribute('href')

        print(f"Job Title: {job_title}")
        print(f"Company: {company}")
        print(f"Location: {location}")
        print(f"Link: {link}")
        print("\n---\n")

    # Close the WebDriver
    driver.quit()

# Example usage for searching Python jobs in Surat, India
scrape_indeed_jobs(query='python', location='surat')
```
